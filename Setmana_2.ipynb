```python
diccionari1 = {"clau1": 1, "clau2": 2, "clau3": 3}
print(diccionari1)
diccionari2 = {1:"hola", 2:"adeu", 3: "Python"}
print(diccionari2)
print(diccionari1["clau2"])
#Afegir una clau al diccionari
diccionari1["clau4"]=4
print(diccionari1)
#Eliminar una clau del diccionari
del diccionari1["clau1"]
print(diccionari1)
#Actualitzar un valor d'una clau
diccionari1["clau2"] = 100
print(diccionari1)
#Comprobar si una clau existeix
print("clau3" in diccionari1)
print("clau5" not in diccionari1)
# Obtenir totes les claus
print(diccionari1.keys())
# Obtenir tots els valors
print(diccionari1.values())
#Obtenir tots els elements del diccionari
print(diccionari1.items())
#Buidar el diccionari
diccionari1.clear()
print(diccionari1)
#Crear un diccionari buit
diccionari3 = {}
print(diccionari3)
#Copiar un diccionari
diccionari4 = diccionari2.copy()
print(diccionari4)
#Altres m√®todes
diccionari4.get(2)
diccionari4.pop(2) #treu l'element amb la clau 2
print(diccionari4)
diccionari4.popitem() #treu l'ultim element del diccionari
print(diccionari4)
diccionari4.update({1:10, 2: 20})
print(diccionari4)
# Recorrer un diccionari
for key in diccionari2:
  print(key, diccionari2[key])
for key, value in diccionari2.items():
  print(key, value)
#Diccionaris dins de diccionaris
diccionari5 = {"a": 1, "b":{"x":100, "y":200}}
print(diccionari5)
print(diccionari5["b"]["y"])
#Afegir un diccionari a un diccionari existent
diccionari5["b"].update({"z":300})
print(diccionari5)
#Diccionaris dins de llistes
llista_diccionaris = [{"a":1},{"b":2}, {"c":3}]
print(llista_diccionaris[0])
print(llista_diccionaris[0]["a"])
#Llistes dins de diccionaris
diccionari_llistes = {"a":[1,2,3], "b":[4,5,6]}
print(diccionari_llistes["a"][0])
print(diccionari_llistes["b"][1])
```
