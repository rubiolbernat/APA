## Emulación de MATLAB usando Python

Como MATLAB, Python es un lenguaje interpretado, lo cual quiere decir que las líneas de código se ejecutan conforme
el intérprete las lee. Esto puede hacerse de dos maneras: escribiéndolas directamente en un interfaz de usuario, como
pueda ser, en el caso del Python, el IDLE o el ipython, o escribiéndolas en un fichero de texto (habitualmente, con la
extensión `.py`) y pasánselo como argumento al propio `python`.

Aunque ambos lenguajes son exasperantemente lentos, su potencia radica en que las operaciones más intensivas en términos
de recursos de cálculo pueden ser implementadas en lenguajes compilados mucho más eficientes (a menudo, Fortran o C).

A partir de ahí, son más las diferencias entre MATLAB y Python que las semejanzas:

- Python es un lenguaje de programación en cuyo diseño intervino el propósito explícito de que fuera *bonito* y ayudara a hacer
  programas *bonitos*. La filosofía del Python, que puede consultar en el [Zen de Python](https://peps.python.org/pep-0020/),
  enumera toda una serie de principios (aforismos, los llaman ellos), orientadas a conseguir este propósito.

  Un ejemplo de la adherencia a estos principios de *belleza* es que, cuando se constató que la primera versión realmente
  operativa del lenguaje, Python 2, incumplía muchos de ellos, la decisión fue diseñar un lenguaje nuevo de trinca, muy
  parecido al original, pero incompatible con él, el Python 3.

  Por su parte, MATLAB, fue concebido como un *MATrix LABoratory*, en el que el manejo de matrices fuera lo más sencillo,
  versátil y potente posible. A partir de ahí, MATLAB fue evolucionando y creciendo y creciendo todavía más. Casi siempre
  siguiendo la política de *compatibilidad hacia atrás*, que tantas desgracias ha ocasionado.

- MATLAB es un software propietario que, en su versión estándar y en otoño de
  2024, tiene un precio de 2.250 €, con licencia permanente, o 900 €, con una licencia de un año. Pero esa versión estándar
  sólo incluye el intérprete de MATLAB, sus sistemas de gráficos e interfaz gráfico de usuario, y poca cosa más. Cuando se
  desea usar MATLAB para algo interesante, es necesario adquirir separadamente alguna de sus *toolboxes*, y éstas pueden ser
  más caras incluso que el propio MATLAB.

  Python es de código abierto y gratuito, y buena parte de las aplicaciones que proporcionan las toolboxes de MATLAB están
  disponibles en paquetes desarrollados por una amplia comunidad de programadores de todo el mundo.

  - Una toolbox de MATLAB que, por ahora, no tiene equivalente enteramente satisfactorio en Python es *simulink*. Esta
    toolbox proporciona un entorno gráfico de simulación y modelado de sistemas dinámicos, y hay gente que considera que
    es el único motivo para seguir usando MATLAB en lugar de Python.

    Ahora bien, el precio de simulink es prohibitivo: sólo el toolbox básico cuesta más que el propio MATLAB (3.400 € o
    1.360 €/año), y sigue sin proporcionar casi nada, sólo el motor de simulación. Si, por poner un ejemplo, el usuario
    desea generar código C/C++ a partir de simulaciones en tiempo real, el coste total se va a 18.200 € o 7.280 €/año. Todo
    esto, por supuesto, para una sola licencia...
>
- En MATLAB, todas las funciones disponibles en los toolboxes instalados en el ordenador están disponibles directamente.
  En Python, aparte de ser preciso que el paquete concreto esté instalado, habrá que importarlo en el código del programa.

  - De hecho, el mecanismo de importación de Python permite solventar un problema típico de MATLAB: saber qué versión de
    una función se está usando. En MATLAB es habitual que distintos toolboxes incluyan funciones con el mismo nombre,
    pero cuyo funcionamiento o, incluso, cometido puede ser diferente. En principio, se usa la versión correspondiente
    al último toolbox instalado.
>
- En MATLAB, los índices de un vector o matriz se escriben entre paréntesis (`'('` y `')'`), en Python se hace entre
  corchetes (`'['` y `']'`).

  Por otro lado:
  - Los índices en Python arrancan en `0`, contrariamente a MATLAB, en el que los índices lo hacen en `1`.
  - El último elemento de un vector de MATLAB se indica con el identificador `end`; en Python se usa en índice `-1`.
>
- El mecanismo de loncha de Python es semejante al usado en MATLAB, pero el orden y significado de los elementos que la
  definen es diferente.

- El operador exponenciación en MATLAB es el circunflejo (`'^'`). En Python, también existe el operador circunflejo, pero su
  propósito es implementar el 'o lógico' excusivo (XOR). Para elevar un número a una potencia deben emplearse dos asteriscos
  (`'**'`). Por ejemplo, `2 ** 3 == 8`.

- En Python, los números complejos deben expresarse en la forma `real + imag j`, donde debe explicitarse tanto la parte real
  como la imaginaria, y la unidad imaginaria se representa con la letra `j`. Así, la raíz cuadrada de `-1` es:
  `(-1) ** 0.5 == 0 + 1j`.

  En MATLAB se puede usar indistintamente la letra `j` o la `i` para la unidad imaginaria y no es necesario indicar la parte
  real cuando es cero, o la imaginaria cuando es uno.

- En MATLAB, el resultado de toda línea no rematada con punto y coma (`;`) se escribe en pantalla. En Python, el resultado
  de la mayoría de las órdenes no se escribe en pantalla, y el punto y coma sólo sirve para escribir distintas órdenes
  en una misma línea.

  Hay que tener en cuenta, no obstante, que en Python el nombre de una variable por sí solo ya es una orden, que escribe
  en pantalla la representación *oficial* del objeto vinculado al nombre. Por ejemplo, si ejecutamos la orden `a = 1`,
  Python asigna silenciosamente el nombre `a` al objeto entero `1`; pero, si a continuación ejecutamos la orden `a`, Python
  escribe en pantalla la representación oficial del objeto `int(1)`:
>  ```ipython
>  [in]: a = 1
>  [in]: a
>  [out]: 1
>  ```
- Los comentarios en Python cualquier cadena de texto que sigue a un carácter almohadilla (`'#`'). En MATLAB el carácter
  que introduce el comentario es el tanto por ciento (`'%'`), que en Python corresponde al operador 'resto de la división
  entera'.

- En Python existen dos tipos de división: la división real, cuyo operador es la barra oblicua (`'/'`), y cuyo resultado
  siempre es real y de tipo real; y la división entera, cuyo operador es la doble barra oblicua (`'//'`), y cuyo resultado
  siempre es entero, pero cuyo tipo puede ser real o entero, en función del tipo de los operandos.

### Lonchas en Python

En los tipos compuestos de Python, a menudo denominados *contenedores*, el acceso a los elementos que los forman se
realiza encerrando entre corchetes (`'['` y `']'`) la clave del elemento. Aunque no siempre es así, por ahora vamos a
considerar que la clave es el índice del elemento en el contenedor. Este es el caso para los tipos en los que tiene
sentido establecer un orden para sus elementos, entre los que se incluyen las `ndarrays` de `numpy`, y a los que
suele referir como *ordinalizables*.

Es importante remarcar que el tipo del índice debe ser siempre **entero** (`int`). En Python no se permite acceder a
los elementos de un contenedor ordinalizable con números reales, aunque su parte fraccionaria sea idénticamente igual
a cero.

Un vector, independientemente de cómo se implemente, sería un ejemplo de contenedor ordinalizable. Para evitar un
exceso de verborrea, usaremos en adelante el nombre `vector` para referirnos a un tipo de datos de estas características;
aunque hay otros que también lo serían, y también hay contenedores no ordinalizables.

Dado un vector de `N` elementos Python permite el acceso de dos modos básicos:

- Indicando índice del elemento en el vector (`vector[índice]`), donde, a diferencia de MATLAB, el primer elemento
  del vector se corresponde con el índice `0`, y todos los índices son no negativos.
- Indicando el índice desde el último elemento del vector. En este caso, `vector[-1]` se corresponde con el último
  elemento, el `vector[-2]` con el penúltimo, ..., y `vector[-N]` se corresponde con el primero.

Además de este mecanismo de acceso, Python también permite el acceso a cachos o *lonchas* del vector (en inglés, *slices*).
En general, una loncha se define mediante tres enteros separados por el carácter dos puntos (`:`):

```python
            primero : postúltimo : incremento
```

- `primero` es, como su nombre indica, el primer elemento que formará parte de la loncha.
- `postúltimo` también es lo que su nombre indica, pero hay que ir con cuidado: este elemento no es el último que formará
  parte de la loncha, sino el siguiente al que será el último.
- `incremento` es el incremento entre índices consecutivos de la loncha.

Los tres valores toman valores por defecto (nótese que la descripción empieza por el tercero de ellos, `incremento`,
debido a que el comportamiento de los otros dos depende de él):

- `incremento`: si no se indica, se toma el valor `1`, es decir, todos los elementos desde el `primero` y hasta justo
  antes de `postúltimo` se incluirán en la loncha. Además, si no se indica, puede omitirse el segundo carácter dos
  puntos (`vector[primero : postúltimo :]` es lo mismo que `vector[primero : postúltimo]`).

  Un valor negativo de `incremento` implica que el vector se recorre desde el elemento `primero` hacia atrás.
- `primero`, si no se indica, se toma el inicio del vector. Cuál es el primer elemento depende, a su vez, del signo del
  `incremento`. Si `incremento > 0`, el `inicio` por defecto será el primer elemento, con índice igual a 0; si
  `incremento < 0`, el `inicio` por defecto será el último elemento (el índice de éste puede ser `N-1` o el genérico `-1`).

  Eso sí, aunque se omita el valor de `primero` y tomemos su valor por defecto, el primer carácter dos puntos es necesario
  para indicar que se trata de una loncha.
- `postúltimo`, si no se indica, la loncha abarcará todo lo que pueda del vector. El comportamiento también depende del signo
  de `incremento`: si `incremento > 0`, la loncha irá desde `primero` hasta el final del vector; si `incremento < 0`,
  la loncha irá desde `primero` hasta el inicio del vector.
  - Nótese que una loncha del tipo `vector[: 0: -1]` no incluiría el elemento `vector[0]`, ya que el índice `0` sería
    el *postúltimo* que, por definición, no se incluye en la loncha (es el elemento que va después del último que sí
    se incluye).

Algunos ejemplos de lonchas:

- `vector[::]` (o `vector[:]`): loncha formada por todos los elementos de `vector`.
  - Este es un mecanismo útil para construir una copia del vector. En Python, algo del estilo `vector2 = vector` no
    crea una copia de `vector`, sino que proporciona un nuevo nombre para el mismo; cualquier modificación en `vector2`
    afectará también a los elementos de `vector`, y viceversa. Con la construcción `vector2 = vector[:]` obtenemos un
    nuevo vector, independiente del primero.
  - Hay que ir con cuidado, sin embargo, porque este mecanismo no funciona con las `ndarray` de `numpy`. En `numpy`,
    las lonchas no son copias de la `ndarray` sino *vistas* de la misma (*view*, en inglés). Cualquier modificación
    en una vista afecta a la `ndarray` original y cualquier otra vista de ella.
- `vector[::-1]`: loncha formada por los elementos de `vector` tomados desde el último hasta el primero.
  - También puede escribirse como `vector[-1::-1]`.
  - Sin embargo, `vector[-1:0:-1]` sería distinto, ya que, en este caso la loncha no incluiría el primer elemento de
    `vector`.
- `vector[::2]` (o `vector[0::2]`): loncha formada por los elementos pares de `vector`.

### Ejercicio 6.7: vectores y matrices en `numpy`

A la vista del resultado anterior, podemos concluir que, efectivamente, una lista de Python, aunque sea de números,
no es un vector, y la suma de dos listas está definida como su concatenación.

Para manejar vectores, y también matrices y otras estructuras matemáticas de orden superior, hemos de usar la clase
`ndarray` definida en `numpy` (recuerde, alias `np`).

Una `ndarray` es semejante a las matrices de MATLAB, pero con varias diferencias:

- Las `ndarray` pueden tener un número arbitrario de dimensiones. En MATLAB (cuyo nombre proviene de Matrix Laboratory),
  todo son matrices de dos dimensiones. Incluso los escalares son una matriz, sólo que de una fila y una columna. En
  MATLAB es posible construir objetos de más de dos dimensiones, pero esto sólo se puede hacer como una extensión de los
  elementos que forman la matriz, que, en el fondo, sigue siendo de dos dimensiones.
- Los operadores matemáticos estándar operan matricialmente en MATLAB. En concreto, si `A` y `B` son matrices, en MATLAB
  la multiplicación `A * B` es el producto matricial; si queremos el producto elemento a elemento de las dos matrices
  hemos de usar el operador *punteado*: `A .* B`.

  En `numpy`, las operaciones estándar son elemento a elemento; si
  queremos el producto matricial hemos de emplear el operador arroba: `A @ B`. Y si lo que queremos es la división,
  tendremos que multiplicar matricialmente por la inversa del denominador (que, no será `B ** -1` ya que también el
  operador exponenciación actúa elemento a elemento; y recuerde que la exponenciación se realiza con el operador `**`,
  no, como en MATLAB, con el circunflejo `^`).
- En MATLAB hay un montón de constantes típicas (como $\pi$ o $e$) predefinidas. En `numpy` también, pero hay que acceder
  a ellas a través del paquete: $\mathrm{np.pi}\equiv\pi$,  $\mathrm{np.e}\equiv e$, etc.

La manera más inmediata de crear una `ndarray` es invocando al constructor de la clase `np.ndarray()` pasándole como
argumento un contenedor (probablemente de otros contenedores, y así recursivamente) cuyos elementos finales sean todos
números y cuyas dimensiones sean coherentes (no podemos tener una fila con un número de elementos diferente al resto de
filas, etc.).

### Ejercicio 6.8: resolución de ecuaciones lineales

Un problema clásico en ingeniería y matemáticas en general es la resolución de un sistema de ecuaciones lineales.

Por ejemplo:

$$
    \left\{\begin{array}{rcl}
        a_1 x + b_1 y + c_1 z & = & d_1 \\
        a_2 x + b_2 y + c_2 z & = & d_2 \\
        a_3 x + b_3 y + c_3 z & = & d_3
    \end{array}\right.
$$

Este sistema puede representarse en forma matricial del modo siguiente:

$$
    \left[\begin{array}{ccc} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{array}\right]
    \cdot
    \left[\begin{array}{c} x \\ y \\ z \end{array}\right]
    =
    \left[\begin{array}{c} d_1 \\ d_2 \\ d_3 \end{array}\right]
$$

Es conocido que este sistema puede resolverse usando las ecuaciones siguientes, donde el valor de *i*-ésima variable
se calcula como el cociente del determinante de la matriz de coeficientes con la *i*-ésima columna sustituida por el
vector de coeficientes independientes, dividida por el determinante de la matriz de coeficientes:

$$
    x = \frac{\begin{array}{|ccc|} \mathbf{d_1} & b_1 & c_1 \\ \mathbf{d_2} & b_2 & c_2 \\ \mathbf{d_3} & b_3 & c_3 \end{array}}
             {\begin{array}{|ccc|} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{array}}
    \qquad
    y = \frac{\begin{array}{|ccc|} a_1 & \mathbf{d_1} & c_1 \\ a_2 & \mathbf{d_2} & c_2 \\ a_3 & \mathbf{d_3} & c_3 \end{array}}
             {\begin{array}{|ccc|} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{array}}
    \qquad
    z = \frac{\begin{array}{|ccc|} a_1 & b_1 & \mathbf{d_1} \\ a_2 & b_2 & \mathbf{d_2} \\ a_3 & b_3 & \mathbf{d_3} \end{array}}
             {\begin{array}{|ccc|} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{array}}
$$

Nótese que, en Python, estas operaciones se reducen a calcular los determinantes de matrices en las que una columna ha sido
sustituida por los valores del vector de coeficientes independientes. Para el cálculo del determinante, `numpy` proporciona
la función `np.linalg.det()`.

Por ejemplo, dado el sistema siguiente:

$$
    \left\{\begin{array}{ccccccr}
         x & + & 2y &   &    & = &  0 \\
        2x & - &  y & + & z  & = &  6 \\
         x &   &    & + & 2z & = &  4
    \end{array}\right.
$$

Podemos calcular el valor de $x$ con el código siguiente:

```ipython
[in]: M = np.array([[1, 2, 0], [2, -1, 1], [1, 0, 2]])
[in]: V = np.array([0, 6, 4])
[in]: Mx = M.copy()
[in]: Mx[:, 0] = V[:]
[in]: x = np.linalg.det(Mx) / np.linalg.det(M)
[in]: print(x)
[out]: 2.0000000000000004
```

***

Escriba el código necesario para resolver el siguiente sistema de ecuaciones usando el método explicado,
y escriba en pantalla el valor de la solución $x$, $y$, $z$.

$$
    \left\{\begin{array}{ccccccr}
        2x & + &  y & - &  z & = &  1 \\
         x & - & 2y & + & 2z & = &  3 \\
         x & + &  y & + &  z & = &  6
    \end{array}\right.
$$

Puede comparar su resultado con el proporcionado por la función `np.linalg.solve()`, que hace lo mismo, pero
de manera más sencilla...

````python
# Introduzca el código Python de su respuesta
import numpy as np

M = np.array([[2, 1, -1], [1, -2, 2], [1, 1, 1]])
V = np.array([1, 3, 6])
Mx = M.copy()
Mx[:, 0] = V[:]
x = np.linalg.solve(M, V)
print(x)
````

### Ejercicio 6.10: generación y visualización de funciones de 2 variables

Cuando se desea manejar una función en dos dimensiones, pongamos `x` e `y`, el problema del muestreo se *eleva al
cuadrado*, en el sentido de que ahora las muestras se obtienen en el plano: para cada valor de `x` hemos de considerar
todos los valores de `y`, y viceversa. No basta, por tanto, usar `np.arange()` o `np.linspace()` dos veces. La solución es la
función `np.meshgrid()`, que devuelve las coordenadas `x` e `y` repetidas y reorganizadas de manera que, conjuntamente,
representen todos los puntos de la celosía.

Por ejemplo, suponga que queremos representar la función siguiente para el margen $-3\le x,y \le 3$:

$$
    z(x, y) = \left(1 - x/2 + x^5 + y^3\right)\cdot e^{-(x^2+y^2)}
$$

La representación de funciones de dos variables independientes es un problema no trivial. Intentar una representación
tridimensional, aunque posible (véase, por ejemplo, [3D plotting](https://matplotlib.org/stable/gallery/mplot3d/index.html)),
no siempre proporciona un resultado claro y/o adecuado.

La alternativa más habitual consiste en mostrar los valores de la función sobre un plano bidimensional mediante curvas
de nivel o escalas de color, como la función `plt.contourf()`, cuyo modo de empleo puede consultar en
[contourf](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html).

El código necesario sería:

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

xx, yy = np.meshgrid(x, y)

zz = (1 - xx/2 + xx**5 + yy**3) * np.exp(-xx**2 - yy**2)

plt.contourf(xx, yy, zz)
```

Que da como resultado:

![mesh](data/mesh.png)

***

Represente la función siguiente para el rango de valores $-1\le x,y \le 1$:

$$
    z(x, y) = \cos\left(2\pi\sqrt{5x^2+12y^2}\right)
$$

> El resultado debería parecerse a:
>
> ![ojos](data/ojos.png)

````python
# Introduzca el código Python de su respuesta al ejercicio 6.10
x= np.linspace(-1,1,100 )
y= np.linspace(-1,1,100 )

xx, yy = np.meshgrid(x, y)
zz= np.cos(2*np.pi*np.sqrt(5*xx**2+12*yy**2))
plt.contourf(xx, yy, zz)
````

El primer paso, después de importar el módulo, es definir los símbolos que queremos usar a priori, cada símbolo

[in]: display(raices)
